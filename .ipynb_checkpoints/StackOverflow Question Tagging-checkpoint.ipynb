```python
import numpy as np
import pandas as pd
import tensorflow as tf
import nltk
import pickle
from nltk.corpus import stopwords
from tensorflow.keras import *
```

```python
df = pd.read_pickle('cleaned_data.pkl')
```

```python
df.head()
```

```python
with open('TagIndex.txt','rb') as myfile:
    tag_dic = pickle.load(myfile)
```

```python
tag_dic
```

```python
with open('WordToInt.txt','rb') as myfile:
    vocab = pickle.load(myfile)
```

```python
vocab_size = len(vocab)
tag_num = len(tag_dic)
print("Tag Num: ",tag_num," Vocab Size: ",vocab_size)
```

```python
m = 1759 #length of longest question , excluding stopwords
s = set(stopwords.words('english'))
```

```python
s
```

```python
df1 = df.sample(n=1000,random_state=1)
print('Sampling Done!')
```

```python
def tokenize(text):
    l = []
    for word in text.strip().split(' '):
        st = str(word).lower()
        if st not in s:
            if st not in vocab:
                l.append(0)
            else:
                l.append(vocab[st])
    while len(l)<m:
        l.append(0)
    return np.array(l)

def tokenize_tags(text):
    l = np.zeros((1,tag_num),dtype=np.float32)
    for word in text.strip().split(' '):
        l[0,tag_dic[word]] = 1.0
    return l
```

```python
num_iter = 10
model.add(layers.Dense(20000,activation='relu',input_shape=(1759,)))
model.add(layers.Dropout(0.3,noise_shape=None,seed=None))
model.add(layers.Dense(17998,activation='sigmoid'))    
```
